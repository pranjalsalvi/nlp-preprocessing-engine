TASK 1: Conceptual Answers

1. “Love” vs “love” in NLP
- “Love” and “love” are treated differently in raw text because of case sensitivity.
- In NLP, we usually convert both to lowercase so they become the same token: love.
- Otherwise, model may treat them as two different words → increases vocabulary size unnecessarily.

2. What happens if stopwords are not removed?
- Text becomes noisy.
- Model focuses on unimportant words like “is”, “the”, “am”.
- Reduces accuracy in classification, sentiment analysis, etc.

3. When removing stopwords can be harmful?
- Sentiment analysis:
- “I am not happy” → removing “not” changes meaning.
- Question detection:
- “What is this?” → removing “what” breaks intent understanding.

4. Stemming vs Lemmatization
- Stemming: chops words crudely  
  → “playing” → “play” / “studies” → “studi”
- Lemmatization: uses grammar + dictionary  
  → “better” → “good”

TASK 2: ADVANCED PREPROCESSING FUNCTION

```python
import re

def preprocess_text(text):
    if not isinstance(text, str):
        return []

    # Handle empty or useless input
    text = text.strip()
    if len(text) == 0:
        return []

    # Lowercase
    text = text.lower()

    # Remove URLs
    text = re.sub(r'http\S+|www\.\S+', '', text)

    # Remove emails
    text = re.sub(r'\S+@\S+', '', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Remove emojis (basic pattern)
    text = re.sub(r'[^\w\s\']', ' ', text)

    # Handle repeated characters (soooo → soo, goooood → good)
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    tokens = text.split()

    cleaned_tokens = []
    for token in tokens:
        if len(token) <= 2 and token not in ["no", "not", "ok"]:
            continue
        cleaned_tokens.append(token)

    return cleaned_tokens
```

```python
# TASK 3: STRESS TESTING
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

for s in test_sentences:
    tokens = preprocess_text(s)
    print("Original:", s)
    print("Tokens:", tokens)
    print("Sentence:", " ".join(tokens))
    print("-"*50)
```

```python
# TASK 4: TOKEN ANALYTICS
def analyze_tokens(tokens):
    if not tokens:
        return {
            "total_tokens": 0,
            "unique_tokens": 0,
            "avg_token_length": 0
        }

    total = len(tokens)
    unique = len(set(tokens))
    avg_len = sum(len(t) for t in tokens) / total

    return {
        "total_tokens": total,
        "unique_tokens": unique,
        "avg_token_length": round(avg_len, 2)
    }
```

```python
for s in test_sentences:
    tokens = preprocess_text(s)
    stats = analyze_tokens(tokens)

    print("\nSentence:", s)
    print("Stats:", stats)
```

```python
# TASK 5: FREQUENCY ANALYSIS
from collections import Counter

all_tokens = []

for s in test_sentences:
    all_tokens.extend(preprocess_text(s))

counter = Counter(all_tokens)

print("Top 10 most common words:")
print(counter.most_common(10))

print("\nTop 5 least common words:")
print(counter.most_common()[:-6:-1])
```

```python
# TASK 6: FULL PIPELINE
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(" ".join(tokens))

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
result = full_pipeline(test_sentences)

print(result["tokens"])
print(result["clean_sentences"])
```
